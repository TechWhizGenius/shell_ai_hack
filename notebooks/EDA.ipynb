```python
print("Missing in train:")
print(train_df.isnull().sum().loc[lambda x: x > 0])

print("Missing in test:")
print(test_df.isnull().sum().loc[lambda x: x > 0])
```

```python
train_ranges = train_df.agg(['min', 'max']).T
train_ranges.columns = ['Min', 'Max']
train_ranges
```

```python
test_ranges = test_df.agg(['min', 'max']).T
test_ranges.columns = ['Min', 'Max']
test_ranges
```

```python
# Check for missing values
null_counts = train_df.isnull().sum()
null_counts = null_counts[null_counts > 0]  # filter non-zero nulls

if null_counts.empty:
    print("No missing values in train dataset.")
else:
    print("Columns with missing values:")
    print(null_counts)
```

```python
test_nulls = test_df.isnull().sum()
test_nulls = test_nulls[test_nulls > 0]

if test_nulls.empty:
    print("No missing values in test dataset.")
else:
    print("Columns with missing values:")
    print(test_nulls)
```

From the above code, there are no null vlaues - so, no need to handle missing values.

The features are already standardized (range = (-3,3))
All features are numeric. so, categorical encoding is not needed
